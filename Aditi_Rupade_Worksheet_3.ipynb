## 1
Consider a simple harmonic oscillator with Hamiltonian $$ \hat{H} = \hbar\omega_0 \hat{a}^\dagger\hat{a}.$$ 

(a) Write down the Heisenberg equation for the annihilation operator $\hat{a}$ and show that its solution is $$\hat{a}(t) = {\rm e}^{-{\rm i}\omega_0 t} \hat{a}(0).$$

#### 1.a) Solution-

Given a simple Harmonic oscillator with Hamiltonian
\begin{equation*}
\hat{H} =\hbar \omega _{0}\hat{a}^{\dagger }\hat{a}
\end{equation*}
The Heisenberg equation of of the annihilation operator is given as
\begin{gather*}
\frac{d\hat{a}}{dt} =\frac{i}{\hbar }[\hat{H} ,\hat{a}]\\
\frac{d\hat{a}}{dt} =\frac{i}{\hbar }\left[ \hbar \omega _{0}\hat{a}^{\dagger }\hat{a} ,\hat{a}\right]\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left[\hat{a}^{\dagger }\hat{a} ,\hat{a}\right]\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }\hat{a}\hat{a} -\hat{a}\hat{a}^{\dagger }\hat{a}\right)\\
Adding\ and\ Subtracting\ \hat{a}^{\dagger }\hat{a}\hat{a}\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }\hat{a}\hat{a} -\hat{a}^{\dagger }\hat{a}\hat{a} +\hat{a}^{\dagger }\hat{a}\hat{a} -\hat{a}\hat{a}^{\dagger }\hat{a}\right)\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }(\hat{a}\hat{a} -\hat{a}\hat{a}) +\left(\hat{a}^{\dagger }\hat{a} -\hat{a}\hat{a}^{\dagger }\right)\hat{a}\right)\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }[\hat{a} ,\hat{a}] +\left[\hat{a}^{\dagger } ,\hat{a}\right]\hat{a}\right)\\
\frac{d\hat{a}}{dt} =i\omega _{0}( 0-\hat{a})\\
\frac{d\hat{a}}{dt} =-i\omega _{0}\hat{a}( t)
\end{gather*}
Let $\displaystyle \hat{a}( t)$ be equal to 
\begin{gather*}
\hat{a}( t) =e^{-i\omega _{0} t}\hat{a}( 0)\\
\frac{d\hat{a}}{dt} =\frac{d}{dt} e^{-i\omega _{0} t}\hat{a}( 0)\\
\frac{d\hat{a}}{dt} =-i\omega _{0}\left( e^{-i\omega _{0} t}\hat{a}( 0)\right)\\
\frac{d\hat{a}}{dt} =-i\omega _{0}\hat{a}( t)
\end{gather*}

The solution of the annihilation operator is
\begin{equation*}
\hat{a}( t) =e^{-i\omega _{0} t}\hat{a}( 0)
\end{equation*}

#### 2.a) Solution-

The interaction picture Hamiltonian (as shown in previous problem set) is derived as
\begin{gather}
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }(\hat{H} -\hat{H}_{0}) e^{i\hat{H}_{o} t/\hbar } \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar \omega _{0}\hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right) -\hbar \omega _{d}\hat{a}^{\dagger }\hat{a}\right) e^{i\hat{H}_{o} t/\hbar } \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar ( \omega _{0} -\omega _{d})\hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right)\right) e^{i\hat{H}_{o} t/\hbar } \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar \Delta \hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right)\right) e^{i\hat{H}_{o} t/\hbar } \notag\\
where\ \Delta =\omega _{0} -\omega _{d} \notag\\
 \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar \Delta \hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right)\right) e^{i\hat{H}_{o} t/\hbar }      ...(1)
\end{gather}
Solving equation (1) in parts,
\begin{equation*}
e^{-i\hat{H}_{o} t/\hbar } \hbar \Delta \hat{a}^{\dagger }\hat{a} e^{i\hat{H}_{o} t/\hbar } =\hbar \Delta \left( 1-i\omega _{d}\hat{a}^{\dagger }\hat{a} t...\right)\hat{a}^{\dagger }\hat{a}\left( 1+i\omega _{d}\hat{a}^{\dagger }\hat{a} t...\right)
\end{equation*}
Using the property $\displaystyle \left[\hat{a}^{\dagger }\hat{a} ,\hat{a}^{\dagger }\hat{a}\right] =0$, all terms will become zero except first term, so we will end up with
\begin{equation*}
e^{-i\hat{H}_{o} t/\hbar } \hbar \Delta \hat{a}^{\dagger }\hat{a} e^{i\hat{H}_{o} t/\hbar } =\hbar \Delta \hat{a}^{\dagger }\hat{a}
\end{equation*}
Solving for the next part
\begin{equation*}
e^{-i\hat{H}_{o} t/\hbar } \hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right) e^{i\hat{H}_{o} t/\hbar } =\hbar Fcos( \omega _{d} t+\varphi )\left( e^{-i\hat{H}_{o} t/\hbar }\hat{a} e^{i\hat{H}_{o} t/\hbar } +e^{-i\hat{H}_{o} t/\hbar }\hat{a}^{\dagger } e^{i\hat{H}_{o} t/\hbar }\right)
\end{equation*}
Using the property $\displaystyle \left[\hat{a}^{\dagger }\hat{a} ,\hat{a}\right] =-\hat{a} \ ,\left[\hat{a}^{\dagger }\hat{a} ,\hat{a}^{\dagger }\right] =\hat{a}^{\dagger } \ $, the terms reduce as follows
\begin{gather*}
=\frac{\hbar F}{2} cos( \omega _{d} t+\varphi )\left( e^{-i\omega _{d} t}\hat{a} +e^{i\omega _{d} t}\hat{a}^{\dagger }\right)\\
=\frac{\hbar F}{2}\left( e^{-i( \omega _{d} t+\varphi )} +e^{i( \omega _{d} t+\varphi )}\right)\left( e^{-i\omega _{d} t}\hat{a} +e^{i\omega _{d} t}\hat{a}^{\dagger }\right)\\
=\frac{\hbar F}{2}\left( e^{i\varphi }\hat{a} +e^{-i\varphi }\hat{a}^{\dagger } +e^{-i( 2\omega _{d} t+\varphi )}\hat{a} +e^{i( 2\omega _{d} t+\varphi )}\hat{a}^{\dagger }\right)
\end{gather*}
Ignoring the last two terms as RWA, we get
\begin{equation*}
=\frac{\hbar F}{2}\left( e^{i\varphi }\hat{a} +e^{-i\varphi }\hat{a}^{\dagger }\right)
\end{equation*}
Substituting the reduced values in (1)
\begin{equation*}
\hat{H}_{I} =\hbar \Delta \hat{a}^{\dagger }\hat{a} +\frac{\hbar F}{2}\left( e^{i\varphi }\hat{a} +e^{-i\varphi }\hat{a}^{\dagger }\right)
\end{equation*}

4.a) Solution-

Given that $\displaystyle \ket{\psi _{+}}$ is normalised, the value of $\displaystyle N_{+}$ can be calculated as
\begin{gather}
|\braket{\psi _{+}}{\psi _{+}} |^{2} =1 \notag\\
|\frac{1}{N_{+}}\left(\bra{\overline{\alpha }} +\bra{-\overline{\alpha }}\right)\left(\ket{\overline{\alpha }} +\ket{-\overline{\alpha }}\right) |^{2} =1 \notag\\
|N_{+} |^{2} =|\braket{\overline{\alpha }}{\overline{\alpha }} +\braket{\overline{\alpha }}{-\overline{\alpha }} +\braket{-\overline{\alpha }}{\overline{\alpha }} +\braket{-\overline{\alpha }}{-\overline{\alpha }} |^{2} \ \ ...( 5)
\end{gather}
Using the property $\displaystyle \braket{\beta}{\alpha } =exp\left( -\frac{1}{2}\left( |\alpha |^{2} +|\beta |^{2} -2\beta ^{*} \alpha \right)\right)$. Solving in parts
\begin{gather*}
\braket{\overline{\alpha }}{\overline{\alpha }} =exp\left( -\frac{1}{2}\left( |\overline{\alpha } |^{2} +|\overline{\alpha } |^{2} -2\overline{\alpha }^{*}\overline{\alpha }\right)\right) =exp\left( -\frac{1}{2}\left( 2|\overline{\alpha } |^{2} -2|\overline{\alpha } |^{2}\right)\right) =exp( 0) =1\\

\braket{-\overline{\alpha }}{-\overline{\alpha }} =exp\left( -\frac{1}{2}\left( |-\overline{\alpha } |^{2} +|-\overline{\alpha } |^{2} -2( -\overline{\alpha })^{*}( -\overline{\alpha })\right)\right) =exp\left( -\frac{1}{2}\left( 2|\overline{\alpha } |^{2} -2|\overline{\alpha } |^{2}\right)\right) =exp( 0) =1
\end{gather*}
Substituting in eqn (5)
\begin{gather*}
|N_{+} |^{2} =|\braket{\overline{\alpha }}{\overline{\alpha }} +\braket{\overline{\alpha }}{-\overline{\alpha }} +\braket{-\overline{\alpha }}{\overline{\alpha }} +\braket{-\overline{\alpha }}{-\overline{\alpha }} |^{2}\\
|N_{+} |^{2} =|1+exp\left( -|\overline{\alpha } |^{2}\right) +exp\left( -|\overline{\alpha } |^{2}\right) +1|^{2}\\
N_{+} =2+2exp\left( -2|\overline{\alpha } |^{2}\right)\\
N_{+} =2+2e^{-2|\overline{\alpha } |^{2}}
\end{gather*}

(b) Derive an expression for the Husimi Q-function for the state $|\psi_+\rangle$. Make a [contour plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html) of the Q-function in the complex $\alpha$-plane for $\bar{\alpha} = 1$ and $\bar{\alpha} = 3$.

#### 4.b) Solution-

The Husumi Q-function for $\displaystyle \ket{\psi _{+}}$ is calculated as
\begin{gather*}
Q=\frac{1}{\pi } |\braket{\alpha }{\psi _{+}} |^{2}\\
Q=\frac{1}{\pi N_{+}^{2}} |\braket{\alpha }{\overline{\alpha }} +\braket{\alpha }{-\overline{\alpha }} |^{2}\\
Q=\frac{1}{\pi N_{+}^{2}} |exp\left( -\frac{1}{2}\left( |\alpha |^{2} +|\overline{\alpha } |^{2} -2\alpha ^{*}\overline{\alpha }\right)\right) +exp\left( -\frac{1}{2}\left( |\alpha |^{2} +|-\overline{\alpha } |^{2} -2\alpha ^{*}( -\overline{\alpha })\right)\right) |^{2}\\
Q=\frac{1}{\pi N_{+}^{2}} |exp\left( -\frac{1}{2}\left( |\alpha |^{2} +|\overline{\alpha } |^{2}\right)\right)\left[ exp\left( \alpha ^{*}\overline{\alpha }\right) +exp\left( -\alpha ^{*}\overline{\alpha }\right)\right] |^{2}\\
Q=\frac{e^{-\left( |\alpha |^{2} +|\overline{\alpha } |^{2}\right)}}{\pi N_{+}^{2}} |e^{\alpha ^{*}\overline{\alpha }} +e^{-\alpha ^{*}\overline{\alpha }} |^{2}
\end{gather*}

```python
def Husumi_Q_func(alpha_bar,alpha):
    n_plus=2+2*(np.exp(-2*(np.abs(alpha_bar)**2)))
    a=np.exp(-((np.abs(alpha)**2) + (np.abs(alpha_bar)**2)))
    b=np.pi*(n_plus**2)
    c=np.abs((np.exp(np.conj(alpha)*alpha_bar))+(np.exp(-np.conj(alpha)*alpha_bar)))**2
    return ((a/b)*c)    
```

```python
def plot_Q_func(alpha_bar):
    x=np.linspace(-5,5,100)
    y=np.linspace(-5,5,100)
    Q=np.zeros([100,100])
    
    for i in range(len(x)):
        for j in range(len(y)):
            Q[j,i]=Husumi_Q_func(alpha_bar,x[i]+1j*y[j])
            
    _,ax = plt.subplots(figsize=(8,8))
    plt.contourf(x,y,Q)
    ax.set_title('Q-function for $\\bar{\\alpha}=$%s'% alpha_bar)
    ax.set_xlabel('$Re \\alpha$')
    ax.set_ylabel('$Im \\alpha$')
```

#### 4.c) Solution-

Given the mixed state $\displaystyle \hat{\rho }$ is normalised, we can express the state as
\begin{gather*}
tr(\hat{\rho }) =1\\
tr\left[\frac{1}{N_{0}}\left(\ket{\overline{\alpha }}\bra{\overline{\alpha }} +\ket{-\overline{\alpha }}\bra{-\overline{\alpha }}\right)\right] =1\\
\frac{1}{N_{0}}\left(\braket{\overline{\alpha }}{\overline{\alpha }} +\braket{-\overline{\alpha }}{-\overline{\alpha }}\right) =1\\
N_{0} =\braket{\overline{\alpha }}{\overline{\alpha }} +\braket{-\overline{\alpha }}{-\overline{\alpha }}
\end{gather*}
Using results from 4.a
\begin{gather*}
\braket{\overline{\alpha }}{\overline{\alpha }} =1\ ,\ \braket{-\overline{\alpha }}{-\overline{\alpha }} =1\\
N_{0} =2
\end{gather*}
The mixed state is $\displaystyle \hat{\rho } =\frac{1}{2}\left(\ket{\overline{\alpha }}\bra{\overline{\alpha }} +\ket{-\overline{\alpha }}\bra{-\overline{\alpha }}\right)$. The Q-function for this is derived using
\begin{gather}
Q( \alpha ) =\frac{1}{\pi }\left(\bra{\alpha }\hat{\rho }\ket{\alpha }\right) \notag\\
=\frac{1}{2\pi }\left(\bra{\alpha }\left(\ket{\overline{\alpha }}\bra{\overline{\alpha }} +\ket{-\overline{\alpha }}\bra{-\overline{\alpha }}\right)\ket{\alpha }\right) \notag\\
=\frac{1}{2\pi }\left(\braket{\alpha }{\overline{\alpha }}\braket{\overline{\alpha }}{\alpha } +\braket{\alpha }{-\overline{\alpha }}\braket{-\overline{\alpha }}{\alpha }\right) \ \ ...\ ( 6)
\end{gather}
Solving eqn (6) in parts
\begin{gather*}
\braket{\alpha }{\overline{\alpha }}\braket{\overline{\alpha }}{\alpha } =exp\left( -\frac{1}{2}\left( |\alpha |^{2} +|\overline{\alpha |}^{2} -2\alpha ^{*}\overline{\alpha }\right)\right) exp\left( -\frac{1}{2}\left( |\overline{\alpha |}^{2} +|\alpha |^{2} -2\overline{\alpha }^{*} \alpha \right)\right)\\
\braket{\alpha }{\overline{\alpha }}\braket{\overline{\alpha }}{\alpha } =exp\left( -\left( |\alpha |^{2} +|\overline{\alpha |}^{2} -\alpha ^{*}\overline{\alpha } -\overline{\alpha }^{*} \alpha \right)\right)\\
\braket{\alpha }{\overline{\alpha }}\braket{\overline{\alpha }}{\alpha } =exp\left( -|\alpha -\overline{\alpha } |^{2}\right)\\
\\
\braket{\alpha }{-\overline{\alpha }}\braket{-\overline{\alpha }}{\alpha } =exp\left( -\frac{1}{2}\left( |\alpha |^{2} +|-\overline{\alpha |}^{2} -2\alpha ^{*}( -\overline{\alpha })\right)\right) exp\left( -\frac{1}{2}\left( |-\overline{\alpha |}^{2} +|\alpha |^{2} -2( -\overline{\alpha })^{*} \alpha \right)\right)\\

\braket{\alpha }{-\overline{\alpha }}\braket{-\overline{\alpha }}{\alpha } =exp\left( -|\alpha +\overline{\alpha } |^{2}\right)
\end{gather*}
Substituting in (6)
\begin{gather*}
Q( \alpha ) =\frac{1}{2\pi }\left(\braket{\alpha }{\overline{\alpha }}\braket{\overline{\alpha }}{\alpha } +\braket{\alpha }{-\overline{\alpha }}\braket{-\overline{\alpha }}{\alpha }\right)\\
Q( \alpha ) =\frac{e^{-|\alpha -\overline{\alpha } |^{2}} +e^{-|\alpha +\overline{\alpha } |^{2}}}{2\pi }
\end{gather*}

The harmonic oscillator is infinite-dimensional and therefore its states and operators cannot be exactly represented by any numerical array. However, one can find an approximate representation by *truncation* of the Hilbert space to a finite size.
    
(a) Let $$\hat{a} = \sum_{n=1}^{d-1} \sqrt{n}|n-1\rangle\langle n|$$ be the approximate annihilation operator in the $d$-dimensional truncated Hilbert space. (This becomes the exact annihilation operator as $d\to\infty$.) For finite $d$, find an analytical expression for the operator $\hat{a}^\dagger \hat{a}$ and compute the commutator $[\hat{a},\hat{a}^\dagger]$.

#### 5.a) Solution-

Analytical expression for $\displaystyle \hat{a}^{\dagger }\hat{a}$ is
\begin{gather*}
\hat{a}^{\dagger }\hat{a} =\sum _{m=1}^{d-1}\sqrt{m}\ket{m}\bra{m-1}\sum _{n=1}^{d-1}\sqrt{n}\ket{n-1}\bra{n}\\
\hat{a}^{\dagger }\hat{a} =\sum _{m=1}^{d-1}\sum _{n=1}^{d-1}\sqrt{m}\sqrt{n}\ket{m}\bra{m-1}\ket{n-1}\bra{n}\\
\hat{a}^{\dagger }\hat{a} =\sum _{m=1}^{d-1}\sum _{n=1}^{d-1}\sqrt{m}\sqrt{n}\ket{m}\bra{n} \delta _{mn}\\
\hat{a}^{\dagger }\hat{a} =\sum _{n=1}^{d-1} n\ket{n}\bra{n}
\end{gather*}
Similarly calculating $\displaystyle \hat{a}\hat{a}^{\dagger }$
\begin{gather*}
\hat{a}\hat{a}^{\dagger } =\sum _{n=1}^{d-1}\sqrt{n}\ket{n-1}\bra{n}\sum _{m=1}^{d-1}\sqrt{m}\ket{m}\bra{m-1}\\
\hat{a}\hat{a}^{\dagger } =\sum _{n=1}^{d-1}\sum _{m=1}^{d-1}\sqrt{m}\sqrt{n}\ket{m-1}\bra{m}\ket{n}\bra{n-1}\\
\hat{a}\hat{a}^{\dagger } =\sum _{m=1}^{d-1}\sum _{n=1}^{d-1}\sqrt{m}\sqrt{n}\ket{m-1}\bra{n-1} \delta _{mn}\\
\hat{a}\hat{a}^{\dagger } =\sum _{n=1}^{d-1} n\ket{n-1}\bra{n-1}
\end{gather*}
Computing the commutation relation $\displaystyle \left[\hat{a} ,\hat{a}^{\dagger }\right]$
\begin{gather*}
\left[\hat{a} ,\hat{a}^{\dagger }\right] =\hat{a}\hat{a}^{\dagger } -\hat{a}^{\dagger }\hat{a}\\
\left[\hat{a} ,\hat{a}^{\dagger }\right] =\sum _{n=1}^{d-1} n\ket{n-1}\bra{n-1} -n\ket{n}\bra{n}\\
\left[\hat{a} ,\hat{a}^{\dagger }\right] =1\ket{0}\bra{0} -1\ket{1}\bra{1} +2\ket{1}\bra{1} -2\ket{2}\bra{2} ...( d-1)\ket{d-2}\bra{d-2} -( d-1)\ket{d-1}\bra{d-1}\\
\left[\hat{a} ,\hat{a}^{\dagger }\right] =1\ket{0}\bra{0} +( 2-1)\ket{1}\bra{1} +( 3-2)\ket{2}\bra{2} ...-( d-1)\ket{d-1}\bra{d-1}\\
\left[\hat{a} ,\hat{a}^{\dagger }\right] =1\ket{0}\bra{0} +\ket{1}\bra{1} +\ket{2}\bra{2} ...-( d-1)\ket{d-1}\bra{d-1}\\
\left[\hat{a} ,\hat{a}^{\dagger }\right] =\mathbb{1} -( d-1)\ket{d-1}\bra{d-1}
\end{gather*}
This shows that the commutation relation acts as identity on the n-1 states and acts differently on the nth state.


(b) Define a Python function that generates the $d$-dimensional matrix representation of $\hat{a}$ for any integer $d$. For $d=5$, numerically test your function: check the commutation relation $[\hat{a},\hat{a}^\dagger]$ and confirm that $\hat{a}|4\rangle$ yields the correct result, where $|4\rangle$ is a Fock state. 

```python
def annihilation_matrix(d):
    '''
    Input- d dimension
    Output- matrix representation
    Function- compute matrix of a as a=sqrt(n)|n-1><n|
    '''
    
    annhilation_op=np.zeros((d,d))
    for n in range(d):
        annhilation_op[n-1][n]=np.sqrt(n)
        
    return annhilation_op
```

```python
d=5
annhilation_op=annihilation_matrix(d)
annihilation_op_dagger=annhilation_op.T.conj()
commutation_relation=annhilation_op@annihilation_op_dagger - annihilation_op_dagger@annhilation_op
print('The commutation relation holds where it acts as identity on n-1 states except the nth state\n',commutation_relation)
```

```python
fock_state=np.array([0,0,0,0,1])
result=annhilation_op@fock_state
print('The result of a|4> is confirmed to return as sqrt(4) on 4th state\n',result)
```

```python
def create_displacement_op(alpha,d):
    annhilation_op=annihilation_matrix(d)
    annhilation_op_dagger=annhilation_op.T.conj()
    alpha_conj=np.conj(alpha)
    displacement_op=expm((alpha*annhilation_op_dagger)-(alpha_conj*annhilation_op))
    return displacement_op
```

#### 5.d) Solution-

The Q-function for some arbitrary $\displaystyle \alpha $ is calculated as
\begin{equation*}
Q( \alpha ) =\frac{1}{\pi } |\braket{\hat{D}( \alpha )}{\psi } |^{2}
\end{equation*}

```python
def truncated_space_Q_func(alpha,psi):
    d=len(psi)
    displacement_op=create_displacement_op(alpha,d)
    displacement_psi=displacement_op.T.conj()@psi
    #As this displacement operator is a matrix, selecting the truncated space for 0th dimension
    Q=(np.abs(displacement_psi[0])**2)/np.pi
    return Q
```

$$ \hat{H} = \hbar\omega_0\hat{a}^\dagger\hat{a} + \hbar F\cos(\omega_{\rm d} t)\left(\hat{a}+\hat{a}^\dagger\right).$$

(a) Write a function that generates this Hamiltonian (for any given $t$) in a truncated Hilbert space with $d=16$.

```python
def Generate_Hamiltonian_6a(d,F,w_0,w_d,t):
    annihilation_op=annihilation_matrix(d)
    annihilation_op_dagger=annihilation_op.T.conj()
    H = w_0*(annihilation_op_dagger@annihilation_op)+F*np.cos(w_d*t)*(annihilation_op+annihilation_op_dagger)
    return H
```

(b) Numerically calculate the dynamics of the quantum state in the Schrödinger picture starting from the vacuum state $|\psi(0)\rangle = |0\rangle$, and plot the mean value of the coordinate, $\langle \hat{q}(t)\rangle,$ as a function of time up to $\omega_0 t=100$.  Choose parameters $F=0.1\omega_0$ and $\omega_{\rm d} = 0.99\omega_0$.  (Recall that Worksheet 1 explored techniques to numerically calculate time evolution.)

```python
def Quantum_Dynamics_of_SHO(F,w_0,w_d,total_time,d, create_hamiltonian):
    #Initialize Time interval
    t = np.linspace(0, total_time, total_time*10)
    expectation_value_q = []
    
    #Step 1 - Define q_t=sqrt(hbar/2*d*w_0) (a + a_dagger)
    q_t = ((1/2*d*w_0)**(1/2))*(annihilation_matrix(d) + annihilation_matrix(d).T.conj())
    
    #Step 2 - Intialize vaccum state
    psi_t = create_psi(0,d)
    
    #Iterate over time interval
    for i in range(len(t)):
        #Step 3 - Create Hamiltonian
        H=create_hamiltonian(d,F,w_0,w_d,t[i])
        
        #Step 4 - Create Time Evolution operator
        U = expm(-1j*H*0.1)
        
        #Step 5 - Evolve state over time
        psi_t = U@psi_t
        
        #Step 6 - Calculate <q_t>=<psi|q_t|psi>
        expectation_value_q.append(psi_t.T.conj()@q_t@psi_t)
        
    return t,expectation_value_q
```

Now consider a general $M$-mode quadratic Hamiltonian $$\hat{H} = \sum_{j,k=0}^{M-1} H_{jk}\hat{a}_j^\dagger\hat{a}_k$$ and let us assume for simplicity that $\mathbf{d}(0)=0$. The matrix $\mathbf{H}$ is Hermitian and thus can be diagonalised by a unitary transformation:

$$ \mathbf{U}^\dagger \cdot\mathbf{H}\cdot\mathbf{U}= \hbar \mathbf{\Omega} = \hbar \begin{pmatrix} \omega_0 & 0 & 0 \cdots \\ 
0 & \omega_1 & 0 \cdots \\ 
0 & 0 & \omega_2 \cdots \\ \vdots & \vdots & \vdots \;\ddots \end{pmatrix},$$

where $\hbar\omega_j$ are the eigenvalues of $\mathbf{H}$. Therefore, $\mathbf{H} = \hbar\mathbf{U}\cdot\mathbf{\Omega}\cdot\mathbf{U}^\dagger$ or, in components, 
$$H_{jk} = \sum_{l,m}U_{jl} \hbar\omega_l \delta_{lm}U_{km}^*.$$
By plugging this into the Hamiltonian, one finds
$$\hat{H} = \sum_{j=0}^{M-1} \hbar\omega_j \hat{b}_j^\dagger \hat{b}_j,$$
where $\hat{b}_j = \sum_{k}U_{kj}^*\hat{a}_k$. Since $[\hat{b}_j,\hat{b}_k^\dagger]=\delta_{jk}$, this represents the Hamiltonian as a sum of $M$ independent oscillators.

(a) Suppose that the system is in a thermal state, $\hat{\rho} = \frac{{\rm e}^{-\beta \hat{H}}}{Z}$, with $\beta = 1/k_BT$. Convince yourself that for this state, $$\langle \hat{b}_k\rangle = 0 ,\qquad \langle \hat{b}_j^\dagger \hat{b}_k\rangle = \bar{n}(\omega_k)\delta_{jk},$$ where $\bar{n}(\omega)$ is the Bose-Einstein distribution. Hence, show that the correlation matrix for a thermal state is
$$\mathbf{C} = \left( {\rm e}^{\beta \mathbf{H}} - 1\right)^{-1}.$$
(The correlation matrix is defined as in question 7, i.e. $C_{jk} = \langle\!\langle \hat{a}_k^\dagger \hat{a}_j\rangle\!\rangle = \langle \hat{a}_k^\dagger \hat{a}_j\rangle - \langle \hat{a}_k^\dagger\rangle\langle\hat{a}_j\rangle$.)

#### 8.a) Solution-

Given a thermal state $\displaystyle \hat{\rho } =\frac{e^{-\beta \hat{H}}}{Z}$, where Z=$\displaystyle tr\left( e^{-\beta \hat{H}}\right)$, we have
\begin{gather*}
< \hat{b}_{k}  >=tr_{k}\left(\hat{b}_{k}\frac{e^{-\beta \hat{H}}}{Z}\right) \ \\
< \hat{b}_{k}  >=tr_{k}\left(\hat{b}_{k}\frac{e^{-\beta \sum \hbar \omega _{j} b_{j}^{\dagger } b_{j}}}{Z_{k}}\right) \ \\
< \hat{b}_{k}  >=tr_{k}\left(\hat{b}_{k}\frac{e^{-\beta \hbar \omega _{k} b_{k}^{\dagger } b_{k}}}{Z_{k}}\right) \ tr_{l}\left(\hat{b}_{l}\frac{e^{-\beta \sum \hbar \omega _{l} b_{l}^{\dagger } b_{l}}}{Z_{l}}\right) \ \\
< \hat{b}_{k}  >=tr_{k}\left(\hat{b}_{k}\frac{e^{-\beta \hbar \omega _{k} b_{k}^{\dagger } b_{k}}}{Z_{k}}\right)\\
< \hat{b}_{k}  >=0
\end{gather*}
The above value is zero because the thermal state doesn't have any off-diagonal elements.
\begin{gather*}
For\ j\neq k,\ < \hat{b}_{j}^{\dagger }\hat{b}_{k}  >=\ tr_{j}\left(\hat{b}_{j}\frac{e^{-\beta \hbar \omega _{j} b_{j}^{\dagger } b_{j}}}{Z_{j}}\right) \ tr_{k}\left(\hat{b}_{k}\frac{e^{-\beta \hbar \omega _{k} b_{k}^{\dagger } b_{k}}}{Z_{k}}\right) =0\\
For\ j=k,\ < \hat{b}_{k}^{\dagger }\hat{b}_{k}  >=tr_{k}\left(\hat{b}_{k}\hat{b}_{k}\frac{e^{-\beta \hbar \omega _{k} b_{k}^{\dagger } b_{k}}}{Z_{k}}\right)\\
< \hat{b}_{k}^{\dagger }\hat{b}_{k}  >=\frac{1}{e^{-\beta \hbar \omega _{k}} -1} =\overline{n}( \omega _{k})\\
\end{gather*}
We have the transformation $\displaystyle \hat{b}_{j} =\sum _{k} U_{kj}^{*}\hat{a}_{k}$, we can inverse this to get
\begin{gather*}
\ \hat{a}_{j} =\sum _{k} U_{kj}\hat{b}_{k}\\
< \ \hat{a}_{j}  >=\sum _{k} U_{kj} < \hat{b}_{k}  >=0\\
C_{jk} =< \hat{a}_{k}^{\dagger }\hat{a}_{j}  >-< \hat{a}_{k}^{\dagger }  >< \hat{a}_{j}  >\\
C_{jk} =< \hat{a}_{k}^{\dagger }\hat{a}_{j}  >\\
C_{jk} =\sum _{lm} U_{jl} U_{km}^{*} < \hat{b}_{m}^{\dagger }\hat{b}_{l}  >\\
C_{jk} =\sum _{lm} U_{jl} U_{km}^{*}\frac{\delta _{lm}}{e^{-\beta \hbar \omega _{l}} -1}\\
C_{jk} =U\left( e^{-\beta \hbar \mathbf{\Omega}} -1\right)^{-1} U^{\dagger }\\
C_{jk} =\left( e^{-\beta \hbar U\mathbf{\Omega}U^{\dagger }} -1\right)^{-1}\\
C_{jk} =\left( e^{-\beta H} -1\right)^{-1}
\end{gather*}

(b) Let us now apply these results to simulating the dynamics of a single oscillator system in contact with a large many-mode bath. Consider the Hamiltonian $\hat{H} = \hat{H}_S + \hat{H}_B + \hat{H}_{SB}$, where
\begin{align}
&\hat{H}_S = \hbar\nu_0 \hat{a}_0^\dagger \hat{a}_0 \\,\qquad &\hat{H}_B = \sum_{j=1}^{M-1} \hbar\nu_j \hat{a}_j^\dagger \hat{a}_j + \sum_{j=1}^{M-2} \frac{\hbar W}{2} \left( \hat{a}_j^\dagger\hat{a}_{j+1} + \hat{a}_j\hat{a}_{j+1}^\dagger \right),\\
& \hat{H}_{SB} = \frac{\sqrt{\Gamma W}}{2}\left(\hat{a}_0^\dagger \hat{a}_1 + \hat{a}_0\hat{a}_1^\dagger\right).
\end{align}
This models a bath comprising a one-dimensional chain of oscillators coupled to the system at one of its ends ($j=1$). Write a function that constructs the matrix $\mathbf{H}$ for the *full* system-bath Hamiltonian $\hat{H} = \sum_{j,k}H_{jk}\hat{a}_j^\dagger\hat{a}_k$, taking the parameters $\{\nu_j\}$, $W$, $\Gamma$, and $M$ as input. Write another function that constructs the corresponding thermal correlation matrix $\mathbf{C}$. Test your function by printing the thermal correlation matrix for the parameters $M=4$, $\nu_j = \nu_0 = 2$, $W=0.5$, $\Gamma=0.05$ and $\beta=0.5$.

```python
#Here I have calculated the matrix as follows-
# 1.For diagonal values, the values are vj
# 2. For off-diagonal elements, when j=0, add constant from H_SB , when j=1 to M-2, add constant from H_B

def H_full_system(vj,W,gamma,M):
    H=np.zeros((M,M))
    for i in range(M):
        for k in range(M):
            if i==k:
                H[i][k]=vj[i]
                
    off_diagonal_elements=[]
    #First value from H_SB
    off_diagonal_elements.append(np.sqrt(gamma*W)/2)
    for i in range(1,(M-2)+1):
        #Other values from H_B
        off_diagonal_elements.append(W/2)
        
    H=H+np.diag(off_diagonal_elements,1)
    H=H+np.diag(off_diagonal_elements,-1)
    return H
```

```python
def Correlation_Matrix(vj,W,gamma,M,beta):
    H=H_full_system(vj,W,gamma,M)
    C = expm(beta*H)-np.eye(M)
    C = np.linalg.matrix_power(C,-1)
    return C
```

```python
M=4
vj=np.array([2,2,2,2])
gamma=0.05
W=0.5
beta=0.5
print('The full system bath Hamiltonian is-\n',H_full_system(vj,W,gamma,M))
print('The Correlation matrix is-\n',Correlation_Matrix(vj,W,gamma,M,beta))
```
