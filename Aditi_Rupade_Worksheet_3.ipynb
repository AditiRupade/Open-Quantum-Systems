Consider a simple harmonic oscillator with Hamiltonian $$ \hat{H} = \hbar\omega_0 \hat{a}^\dagger\hat{a}.$$ 

(a) Write down the Heisenberg equation for the annihilation operator $\hat{a}$ and show that its solution is $$\hat{a}(t) = {\rm e}^{-{\rm i}\omega_0 t} \hat{a}(0).$$

#### 1.a) Solution-

Given a simple Harmonic oscillator with Hamiltonian
\begin{equation*}
\hat{H} =\hbar \omega _{0}\hat{a}^{\dagger }\hat{a}
\end{equation*}
The Heisenberg equation of of the annihilation operator is given as
\begin{gather*}
\frac{d\hat{a}}{dt} =\frac{i}{\hbar }[\hat{H} ,\hat{a}]\\
\frac{d\hat{a}}{dt} =\frac{i}{\hbar }\left[ \hbar \omega _{0}\hat{a}^{\dagger }\hat{a} ,\hat{a}\right]\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left[\hat{a}^{\dagger }\hat{a} ,\hat{a}\right]\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }\hat{a}\hat{a} -\hat{a}\hat{a}^{\dagger }\hat{a}\right)\\
Adding\ and\ Subtracting\ \hat{a}^{\dagger }\hat{a}\hat{a}\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }\hat{a}\hat{a} -\hat{a}^{\dagger }\hat{a}\hat{a} +\hat{a}^{\dagger }\hat{a}\hat{a} -\hat{a}\hat{a}^{\dagger }\hat{a}\right)\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }(\hat{a}\hat{a} -\hat{a}\hat{a}) +\left(\hat{a}^{\dagger }\hat{a} -\hat{a}\hat{a}^{\dagger }\right)\hat{a}\right)\\
\frac{d\hat{a}}{dt} =i\omega _{0}\left(\hat{a}^{\dagger }[\hat{a} ,\hat{a}] +\left[\hat{a}^{\dagger } ,\hat{a}\right]\hat{a}\right)\\
\frac{d\hat{a}}{dt} =i\omega _{0}( 0-\hat{a})\\
\frac{d\hat{a}}{dt} =-i\omega _{0}\hat{a}( t)
\end{gather*}
Let $\displaystyle \hat{a}( t)$ be equal to 
\begin{gather*}
\hat{a}( t) =e^{-i\omega _{0} t}\hat{a}( 0)\\
\frac{d\hat{a}}{dt} =\frac{d}{dt} e^{-i\omega _{0} t}\hat{a}( 0)\\
\frac{d\hat{a}}{dt} =-i\omega _{0}\left( e^{-i\omega _{0} t}\hat{a}( 0)\right)\\
\frac{d\hat{a}}{dt} =-i\omega _{0}\hat{a}( t)
\end{gather*}

The solution of the annihilation operator is
\begin{equation*}
\hat{a}( t) =e^{-i\omega _{0} t}\hat{a}( 0)
\end{equation*}

```python
dimensionless_time=np.linspace(0,10,100)   #random values of wt
number_fluctuations=[]    #array to store Δn
position_fluctuations=[]    #array to store Δq/q0

#The expectation values of n is constant, calculating it before to save processing
delta_n=np.sqrt(2.5-(1.5*1.5))

for time in dimensionless_time:
    number_fluctuations.append(delta_n)
    q_t=-np.sqrt(2)*np.cos(time)
    q_t_square=4
    position_fluctuations.append(np.sqrt(q_t_square-(q_t*q_t)))
    
#Plot the expectation value
plt.subplots(figsize=(10,5))
plt.xlabel('Dimensionless time $\omega_0$t')
plt.ylabel('Fluctuations')
plt.title('Number fluctuations ans position fluctuations as function of $\omega_0$t')
plt.plot(dimensionless_time,number_fluctuations, label='$\Delta\hat{n}$')
plt.plot(dimensionless_time,position_fluctuations, label='$\Delta\hat{q}/q_0$')
plt.legend(loc=(1.1,0.8),prop={'size': 15})
plt.show()
```

A driven harmonic oscillator is described by the Hamiltonian
$$ \hat{H} = \hbar\omega_0\hat{a}^\dagger\hat{a} + \hbar F\cos(\omega_{\rm d} t+\varphi)\left(\hat{a}+\hat{a}^\dagger\right),$$
where $F$, $\omega_{\rm d}$, and $\varphi$ denote the strength, the frequency, and the phase of the drive, respectively. 

(a) Move to an interaction picture with respect to the free Hamiltonian $\hat{H}_0 = \hbar\omega_{\rm d}\hat{a}^\dagger\hat{a}$ and show that the interaction-picture Hamiltonian under the rotating-wave approximation takes the form
$$\hat{H}_I = \hbar\Delta \hat{a}^\dagger\hat{a} + \frac{\hbar F}{2}\left({\rm e}^{{\rm i}\varphi} \hat{a}+{\rm e}^{-{\rm i}\varphi} \hat{a}^\dagger \right),$$
where $\Delta =\omega_0-\omega_{\rm d}$. State the conditions for this approximation to be valid.


#### 2.a) Solution-

The interaction picture Hamiltonian (as shown in previous problem set) is derived as
\begin{gather}
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }(\hat{H} -\hat{H}_{0}) e^{i\hat{H}_{o} t/\hbar } \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar \omega _{0}\hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right) -\hbar \omega _{d}\hat{a}^{\dagger }\hat{a}\right) e^{i\hat{H}_{o} t/\hbar } \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar ( \omega _{0} -\omega _{d})\hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right)\right) e^{i\hat{H}_{o} t/\hbar } \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar \Delta \hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right)\right) e^{i\hat{H}_{o} t/\hbar } \notag\\
where\ \Delta =\omega _{0} -\omega _{d} \notag\\
 \notag\\
\hat{H}_{I} =e^{-i\hat{H}_{o} t/\hbar }\left( \hbar \Delta \hat{a}^{\dagger }\hat{a} +\hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right)\right) e^{i\hat{H}_{o} t/\hbar }
\end{gather}
Solving equation (1) in parts,
\begin{equation*}
e^{-i\hat{H}_{o} t/\hbar } \hbar \Delta \hat{a}^{\dagger }\hat{a} e^{i\hat{H}_{o} t/\hbar } =\hbar \Delta \left( 1-i\omega _{d}\hat{a}^{\dagger }\hat{a} t...\right)\hat{a}^{\dagger }\hat{a}\left( 1+i\omega _{d}\hat{a}^{\dagger }\hat{a} t...\right)
\end{equation*}
Using the property $\displaystyle \left[\hat{a}^{\dagger }\hat{a} ,\hat{a}^{\dagger }\hat{a}\right] =0$, all terms will become zero except first term, so we will end up with
\begin{equation*}
e^{-i\hat{H}_{o} t/\hbar } \hbar \Delta \hat{a}^{\dagger }\hat{a} e^{i\hat{H}_{o} t/\hbar } =\hbar \Delta \hat{a}^{\dagger }\hat{a}
\end{equation*}
Solving for the next part
\begin{equation*}
e^{-i\hat{H}_{o} t/\hbar } \hbar Fcos( \omega _{d} t+\varphi )\left(\hat{a} +\hat{a}^{\dagger }\right) e^{i\hat{H}_{o} t/\hbar } =\hbar Fcos( \omega _{d} t+\varphi )\left( e^{-i\hat{H}_{o} t/\hbar }\hat{a} e^{i\hat{H}_{o} t/\hbar } +e^{-i\hat{H}_{o} t/\hbar }\hat{a}^{\dagger } e^{i\hat{H}_{o} t/\hbar }\right)
\end{equation*}
Using the property $\displaystyle \left[\hat{a}^{\dagger }\hat{a} ,\hat{a}\right] =-\hat{a} \ ,\left[\hat{a}^{\dagger }\hat{a} ,\hat{a}^{\dagger }\right] =\hat{a}^{\dagger } \ $, the terms reduce as follows
\begin{gather*}
=\frac{\hbar F}{2} cos( \omega _{d} t+\varphi )\left( e^{-i\omega _{d} t}\hat{a} +e^{i\omega _{d} t}\hat{a}^{\dagger }\right)\\
=\frac{\hbar F}{2}\left( e^{-i( \omega _{d} t+\varphi )} +e^{i( \omega _{d} t+\varphi )}\right)\left( e^{-i\omega _{d} t}\hat{a} +e^{i\omega _{d} t}\hat{a}^{\dagger }\right)\\
=\frac{\hbar F}{2}\left( e^{i\varphi }\hat{a} +e^{-i\varphi }\hat{a}^{\dagger } +e^{-i( 2\omega _{d} t+\varphi )}\hat{a} +e^{i( 2\omega _{d} t+\varphi )}\hat{a}^{\dagger }\right)
\end{gather*}
Ignoring the last two terms as RWA, we get
\begin{equation*}
=\frac{\hbar F}{2}\left( e^{i\varphi }\hat{a} +e^{-i\varphi }\hat{a}^{\dagger }\right)
\end{equation*}
Substituting the reduced values in (1)
\begin{equation*}
\hat{H}_{I} =\hbar \Delta \hat{a}^{\dagger }\hat{a} +\frac{\hbar F}{2}\left( e^{i\varphi }\hat{a} +e^{-i\varphi }\hat{a}^{\dagger }\right)
\end{equation*}

Show that the interaction-picture Hamiltonian is diagonalised by the unitary change of basis $\hat{S}={\rm e}^{\hat{G}}$ with $$\hat{G} = \frac{F}{2\Delta}\left({\rm e}^{{\rm i}\varphi} \hat{a}-{\rm e}^{-{\rm i}\varphi} \hat{a}^\dagger \right).$$ Specifically, find the transformed Hamiltonian $\hat{S}^\dagger \hat{H}_I\hat{S}$ and, therefore, give general expressions for eigenvalues and eigenstates of $\hat{H}_I$.

#### 2.d) Solution-

Given the following values
\begin{gather*}
\hat{S} =e^{\hat{G}} \ ,\ where\ \hat{G} =\frac{F}{2\Delta }\left( e^{i\varphi }\hat{a} -e^{-i\varphi }\hat{a}^{\dagger }\right)\\
\hat{H}_{I} =\hbar \Delta \hat{a}^{\dagger }\hat{a} +\frac{\hbar F}{2}\left(\hat{a} e^{i\varphi } +\hat{a}^{\dagger } e^{-i\varphi }\right)
\end{gather*}
Using BCH formula to first calculate the values $\displaystyle \hat{S}\hat{a}\hat{S}$ and $\displaystyle \hat{S}\hat{a}^{\dagger }\hat{S}$ -
\begin{gather}
\hat{S}\hat{a}\hat{S} =e^{-G}\hat{a} e^{G} \notag\\
=\hat{a} -[\hat{G} ,\hat{a}] +\frac{1}{2}[\hat{G} ,[\hat{G} ,\hat{a}]] -...
\end{gather}
Solving in parts-
\begin{gather*}
[\hat{G} ,\hat{a}] =\frac{F}{2\Delta }\left\{\left( e^{i\varphi }\hat{a}\hat{a} -e^{-i\varphi }\hat{a}^{\dagger } a\right) -\left( e^{i\varphi }\hat{a}\hat{a} -e^{-i\varphi }\hat{a}\hat{a}^{\dagger }\right)\right\}\\
[\hat{G} ,\hat{a}] =\frac{F}{2\Delta }\left( -e^{-i\varphi }\hat{a}^{\dagger } a+e^{-i\varphi }\hat{a}\hat{a}^{\dagger }\right)\\
[\hat{G} ,\hat{a}] =\frac{Fe^{-i\varphi }}{2\Delta }\left[\hat{a} ,\hat{a}^{\dagger }\right]\\
[\hat{G} ,\hat{a}] =\frac{Fe^{-i\varphi }}{2\Delta }\\
\\
[\hat{G} ,[\hat{G} ,\hat{a}]] =\frac{F^{2}}{4\Delta ^{2}}\left\{\left( e^{i\varphi } e^{-i\varphi }\hat{a} -e^{-i\varphi } e^{-i\varphi }\hat{a}^{\dagger }\right) -\left( e^{i\varphi } e^{-i\varphi }\hat{a} -e^{-i\varphi } e^{-i\varphi }\hat{a}^{\dagger }\right)\right\}\\
[\hat{G} ,[\hat{G} ,\hat{a}]] =0
\end{gather*}
Similarly, all the next terms in the expansion are zero. Substituting the values in (2), we get
\begin{equation*}
e^{-G}\hat{a} e^{G} =\hat{a} -\frac{Fe^{-i\varphi }}{2\Delta }
\end{equation*}
Similarly for transpose
\begin{equation*}
e^{-G}\hat{a}^{\dagger } e^{G} =\hat{a}^{\dagger } -\frac{Fe^{i\varphi }}{2\Delta }
\end{equation*}
Solving for Hamiltonian
\begin{gather*}
\hat{S}\hat{H}_{I}\hat{S} =e^{-G}\widehat{H_{I}} e^{G}\\
\hat{S}\hat{H}_{I}\hat{S} =e^{-G}\left( \hbar \Delta \hat{a}^{\dagger }\hat{a} +\frac{\hbar F}{2}\left(\hat{a} e^{i\varphi } +\hat{a}^{\dagger } e^{-i\varphi }\right)\right) e^{G}\\
\hat{S}\hat{H}_{I}\hat{S} =\hbar \Delta \left(\hat{a}^{\dagger } -\frac{Fe^{i\varphi }}{2\Delta }\right)\left(\hat{a} -\frac{Fe^{-i\varphi }}{2\Delta }\right) +\frac{\hbar F}{2}\left\{\left(\hat{a} -\frac{Fe^{-i\varphi }}{2\Delta }\right) e^{i\varphi } +\left(\hat{a}^{\dagger } -\frac{Fe^{i\varphi }}{2\Delta }\right) e^{-i\varphi }\right\}\\
\hat{S}\hat{H}_{I}\hat{S} =\hbar \Delta \left(\hat{a}^{\dagger }\hat{a} -\frac{Fe^{i\varphi }}{2\Delta }\hat{a} -\frac{Fe^{-i\varphi }}{2\Delta }\hat{a}^{\dagger } +\frac{F^{2}}{4\Delta ^{2}} e^{0}\right) +\frac{\hbar F}{2}\left(\hat{a} e^{i\varphi } -\frac{Fe^{0}}{2\Delta } +\hat{a}^{\dagger } e^{-i\varphi } -\frac{Fe^{0}}{2\Delta }\right)\\
\hat{S}\hat{H}_{I}\hat{S} =\hbar \Delta \hat{a}^{\dagger }\hat{a} -\frac{\hbar Fe^{i\varphi }}{2}\hat{a} -\frac{\hbar Fe^{-i\varphi }}{2}\hat{a}^{\dagger } +\frac{\hbar F^{2}}{4\Delta } +\frac{\hbar Fe^{i\varphi }}{2}\hat{a} +\frac{\hbar Fe^{-i\varphi }}{2}\hat{a}^{\dagger } -\frac{\hbar F^{2}}{2\Delta }\\
\hat{S}\hat{H}_{I}\hat{S} =\hbar \Delta \hat{a}^{\dagger }\hat{a} -\frac{\hbar F^{2}}{4\Delta }\\
\end{gather*}

The eigenvalues of of $\displaystyle \hat{H}_{I}$ are given by using Fock states-
\begin{gather*}
\hat{H}_{I}|n\rangle =\hat{S}\hat{H}_{I}\hat{S}|n\rangle\\
\hat{H}_{I}|n\rangle =\left( \hbar \Delta \hat{a}^{\dagger }\hat{a} -\frac{\hbar F^{2}}{4\Delta }\right)|n\rangle\\
using\ \hat{a}^{\dagger }\hat{a}|n\rangle =n|n\rangle\\
\hat{H}_{I}|n\rangle =\left( \hbar \Delta n-\frac{\hbar F^{2}}{4\Delta }\right)|n\rangle
\end{gather*}
So the eigenstates of $\displaystyle H_{I}$ are the transformed Fock states $\displaystyle |E_{n}\rangle =|n\rangle$ with eigenenergies $\displaystyle E_{n} =\left( \hbar \Delta n-\frac{\hbar F^{2}}{4\Delta }\right)$
